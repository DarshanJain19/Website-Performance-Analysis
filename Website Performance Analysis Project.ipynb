# Website Performance Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1- Importing the data

```python
df=pd.read_csv("data-export (1).csv") 
df.head()
```

#Cleaning and Systemizing the data,Since column names are not named, so we have to change it.
Now we can see that the 1st row of data looks like a real column values,
we will use iloc[] function here to extract data from 1st row 

```python
df.columns=df.iloc[0]
df.head()
```

#Here some problem arises, firstly columns_name and 1st row values are same, and columns name are very big.


```python
df=df.drop(index=0).reset_index(drop=True)
df.columns=["Channel group","DateHour","Users","Sessions","Engaged Sessions","Average engagement time per session","Engaged sessions per user","Events per session","Engagement rate","Event count"]
```

```python
df.head()
```

# 2-Checking the Formats of all data

```python
df.info()
```

```python
#now converting data types of each columns to their respective types.
#like datehour is not in "datetime" format
```

```python
df["DateHour"] = pd.to_datetime(df["DateHour"], format="%Y%m%d%H",errors='coerce')
df.head()
```

```python
#also changing types of all other columns that are need to be changed from object to numeric values like int, float
#("Channel group","DateHour")only these two columns will not get changed from object to numeric because one is string and other is in datetime format 
```

```python
numeric_cols=df.columns.drop(["Channel group","DateHour"]) 
df[numeric_cols]=df[numeric_cols].apply(pd.to_numeric,errors='coerce') #'coerce'- Replaces unconvertible values with NaN (Not a Number).
df["Hour"] = df["DateHour"].dt.hour # making new column 'Hour' from "DateHour"  
```

```python
df.head()
```

```python
df.info()
```

```python
# data type of all the columns have been changed to their respective numeric values 
```

```python
df.describe()
```

#ques1-What patterns or trends can you observe in website session and user over time?

# Sessions and user over time

```python
sns.set(style="whitegrid") #used for grid
```

```python
plt.figure(figsize=(10,5))
df.groupby("DateHour")[["Sessions","Users"]].sum().plot(ax=plt.gca())
plt.title("Sessions and Users over time") 
plt.xlabel("DateHour")
plt.ylabel("Count")
plt.show()
```

```python
#or we can use this also- This is pure PANDAS Version.
df.groupby("DateHour")[["Sessions", "Users"]].sum().plot(
    kind="line",
    figsize=(10,5),
    title="Sessions and Users over time",
    xlabel="DateHour",
    ylabel="Count"
)
```

>Conclusion -> The line chart shows user activity patterns over time. Sessions are consistently higher than users, indicating that many users return or initiate multiple sessions. There are noticeable spikes at certain hours, suggesting peak engagement periods. Understanding these time windows can help optimize content release, marketing efforts, or support availability for better user engagement.

# Engaged vs Non-engaged Sessions

```python
session_df = df.groupby("Channel group")[["Sessions","Engaged Sessions"]].sum().reset_index()
session_df["Non-Engaged"] = session_df["Sessions"]-session_df["Engaged Sessions"]
session_df_melted = session_df.melt(id_vars = "Channel group",value_vars = ["Engaged Sessions","Non-Engaged"])#.melt() is used to reshape a DataFrame from wide format to long format.

plt.figure(figsize=(8,5))
sns.barplot(data= session_df_melted,x="Channel group", y="value",hue="variable")
plt.title("Engaged vs Non-Engaged")
plt.xticks(rotation=45)
plt.show()
```

>Conclusion = 1- Organic Social have the most Engaged Sessoins and the most Non-Engaged sessoins
2- Direct Channel is a Channel that has more Non-Engaged sessions compared to Engaged Session, so we need to work more on the Direct channel group

# Traffic by Hour and Channel 

```python
heatmap_data= df.groupby(["Hour","Channel group"])["Sessions"].sum().unstack().fillna(0)

plt.figure(figsize=(12,6))
sns.heatmap(heatmap_data , cmap="YlGnBu", linewidths=.5, annot=True, fmt='.0f')
plt.title("Traffic by hour and Channel")
plt.xlabel("Channel group")
plt.ylabel("Hours Of Day")
plt.show()
```

>Conclusion- 1-Organic Social has the most traffic among all the Channels at 12:00 AM.
2-Direct has the most Traffic at 11PM.
3- Organic Search has the most Traffic at 2PM.
4-Refferal has the most Traffic at 9PM.
And rest three channels has the least traffic throughout the day comparitively to other channels.

# Corelation between Engagement rate and sessions over time

```python
df_plot = df.groupby('DateHour')[["Engagement rate", "Sessions"]].mean().reset_index()
plt.figure(figsize=(10,5))
plt.plot(df_plot["DateHour"],df_plot["Engagement rate"],label="Engagement Rate", color="Green")
plt.plot(df_plot["DateHour"],df_plot["Sessions"],label="Sessions", color="Blue")
plt.title("Engagement rate vs Sessions over time")
plt.xlabel("DateHour")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45) 
plt.show()

```

> Conclusion - The session performance shows noticeable fluctuations over time. While the number of sessions is relatively high, the average engagement time and engagement rate remain low. This suggests that although users are visiting frequently, they are not highly engaged during their sessions.
